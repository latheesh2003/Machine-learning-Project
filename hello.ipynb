```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier  # sklearn libraby is popular library in python and 
#that sklearn package tree is module and decisiontreeclassifier is class
music_data= pd.read_csv('music.csv')
X = music_data.drop('genre', axis=1)
Y=music_data['genre']
#we have to create the instance of decision tree and we have to train it
model=DecisionTreeClassifier()
#method to train the dataset
model.fit(X,Y)
#now we are going to predict
predictions=model.predict([[21,1],[22,0]])
predictions
#so the prediction is completed, so next calculating the accuracy
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier  
from sklearn.model_selection import train_test_split  # in the mode_selection module we use the function called train_test_split
from sklearn.metrics import accuracy_score  # function
music_data= pd.read_csv('music.csv')
X = music_data.drop('genre', axis=1)
Y=music_data['genre']
#after the spliting the dataset then it taken as tuple so 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2) # we use 20% of the data to testing

model=DecisionTreeClassifier()

model.fit(X_train,Y_train)

predictions=model.predict(X_test)
# to measure the accuracy of the dataset
score=accuracy_score(Y_test,predictions)
score
```
